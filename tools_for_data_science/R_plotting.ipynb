## Plotting with R

```r
library(IRdisplay)
library(ggplot2)
# dfube cars vector with 5 values
cars <- c(1,4,6,5,10)
plot(cars,type='o')
title(main='cars vs index')
```

```r
ggplot(mtcars,aes(x=mpg,y=wt))+geom_point()+ggtitle('miles per gallon vs weight') + labs(y='weight',x='miles per gallon')
```

```r
ggplot(mtcars,aes(x=disp,y=mpg))+geom_point()+ggtitle('miles per gallon vs displacement') + labs(y='miles per galon',x='displacement')
```

```r
## create boxplot of the distribution of mpg for the individual engine types
mtcars$vs <- as.factor(mtcars$vs) #turn vs (engine type V-shaped or straight) from integer into categorical type
ggplot(aes(x=vs, y=mpg, fill=vs), data = mtcars)+geom_boxplot(alpha=0.3)+theme(legend.position='none')
```

```r
#create histogram of weight 
ggplot(aes(x=wt),data=mtcars)+geom_histogram(binwidth=0.5)
```
